# Loading and combining AMPS Netcdfs 
### Notebook for testing out methods of getting AMPS data into zarr format in our google bucket: gs://ldeo-glaciology 
We already have many AMPS netcdfs in the googlebucket https://storage.googleapis.com/ldeo-glaciology/AMPS/. 
Our aim to to combine them into an analysis-ready format.
The best option will be going from the NCAR HPC access server (?) directly to boogle bucket (GB). But in the meantime we can try going from local netcdfs to GB. This script does this. 

```python
import xarray as xr
import fsspec
import zarr
import gcsfs
#from itertools import product
#from cftime import DatetimeNoLeap
from tqdm.autonotebook import tqdm
xr.set_options(display_style="html");
```

```python
#fs = gcsfs.GCSFileSystem('rosy-hope-23661', token='/Users/jkingslake/.config/gcloud/legacy_credentials/jk3930@columbia.edu/adc.json')
fs = gcsfs.GCSFileSystem('rosy-hope-23661', token=None)
gcsmap = gcsfs.mapping.GCSMap('gs://ldeo-glaciology', gcs=fs, check=False, create=False)
```

### Look at what's in the google bucket.

```python
NCs = fs.ls('gs://ldeo-glaciology/AMPS/wrfout_d03_20161222_week')
```

```python
url = 'https://storage.googleapis.com/' + NCs[1]
```

```python
#url = 'https://storage.googleapis.com/ldeo-glaciology/AMPS/wrfout_d03_20161222_week/wrfout_d03_2016122200_f006.nc'

url = 'https://storage.googleapis.com/' + NCs[0]
with fsspec.open(url) as fobj:
    dset_all = xr.open_dataset(fobj,chunks={'south_north': 10, 'west_east': 10})

for i in tqdm(range(0, len(Ncs)-1)):
    url = 'https://storage.googleapis.com/' + NCs[i]
    with fsspec.open(url) as fobj:
        dset = xr.open_dataset(fobj,chunks={'south_north': 10, 'west_east': 10})
    dset_all = xr.concat([dset_all,dset],'Time')
    
```

```python
dset_all
```

```python
url = 'https://storage.googleapis.com/ldeo-glaciology/AMPS/wrfout_d03_20161222_week/wrfout_d03_2016122200_f006.nc'

with fsspec.open(url) as fobj:
    ds_f006 = xr.open_dataset(fobj,chunks={'south_north': 10, 'west_east': 10})
```

```python
ds_f006.to_zarr('AMPS.zarr')
```

```python
url = 'https://storage.googleapis.com/ldeo-glaciology/AMPS/wrfout_d03_20161222_week/wrfout_d03_2016122200_f009.nc'

with fsspec.open(url) as fobj:
    ds_f009 = xr.open_dataset(fobj,chunks={'south_north': 10, 'west_east': 10})

```

```python
ds_f009.to_zarr('AMPS.zarr', append_dim='Time')
```

```python
AMPS_merged = xr.concat([ds_f006,ds_f009],'Time')
```

```python
AMPS_merged.nbytes /1e9
```

```python
from dask.diagnostics import ProgressBar

# or distributed.progress when using the distributed scheduler
delayed_obj = AMPS_merged.to_zarr('AMPS_merged2', compute=False)

with ProgressBar():
     results = delayed_obj.compute()
```
